# Washington DC Metro Area Crime Heatmap #

**This package makes plotting data on Google Maps effortless. **
`gmplot` package [crash course](https://pypi.org/project/gmplot/)

```python
import pandas as pd
import numpy as np
import gmplot
from IPython.display import Image
```

```python
working_dir = "/home/lee/Documents/GitHubScripts/hgfs/Datasets for GitHub/kaggle_crime_report/"
df = pd.read_csv(working_dir+"dc_crime_add_vars.csv")
print('Input data shape: {}'.format(df.shape))
print('Preview:\n {}'.format(df.head()))
```

```python
print('Types of offense:\n {}'.format(df['OFFENSE'].unique()))
```

**Download the map for the area with suitable zoom level**

```python
zoom_level = 12
gmap = gmplot.GoogleMapPlotter.from_geocode("Washington DC, US", zoom_level)
```

```python
gmap.heatmap(df[df['OFFENSE']=="SEX ABUSE"]['YBLOCK'], df[df['OFFENSE']=="SEX ABUSE"]['XBLOCK'])
gmap.draw("DC_metro_sex_abuse_heatmap.html")
```

```python
gmap.heatmap(df[df['OFFENSE']=='ROBBERY']['YBLOCK'], df[df['OFFENSE']=='ROBBERY']['XBLOCK'])
gmap.draw("DC_metro_robbery_heatmap.html")
```
